```python
import os
import json
import sys
```

```python
"""
Calibration을 통해 생성된 데이터를 확인함
"""


# Read calib.json file into results
file_path = "calib.json"  # Adjust the path if needed

if os.path.exists(file_path):
    with open(file_path, 'r') as f:
        results = json.load(f)
    print(f"Successfully loaded calibration data from {file_path}")
else:
    print(f"Error: File {file_path} not found")
    results = {}  # Initialize with empty dict if file doesn't exist
```

```python
# Print information about the results dictionary structure
print(f"Number of keys in results: {len(results)}")
print(f"Keys in results: {', '.join(list(results.keys())[:5])}... (showing first 5)")

# Check the structure of the first value
first_key = next(iter(results))
first_value = results[first_key]
print(f"Type of data for key '{first_key}': {type(first_value).__name__}")
print(f"Length of data for key '{first_key}': {len(first_value)}")

if isinstance(first_value, list) and len(first_value) > 0:
    print(f"Type of first inner element: {type(first_value[0]).__name__}")
    print(f"Length of first inner element: {len(first_value[0]) if hasattr(first_value[0], '__len__') else 'N/A'}")
```
